```python
import requests
import json
import yaml
import csv
from datetime import datetime

url = "https://api.orphacode.org/EN/ClinicalEntity"
current_date = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
filename = f"clinical_entity_dataset_{current_date}.csv"

response = requests.get(url)
response.raise_for_status()

# Determine the response format (JSON or YAML)
content_type = response.headers.get("Content-Type")
is_json_response = "application/json" in content_type

if is_json_response:
    data = response.json()
else:
    data = yaml.safe_load(response.content)

# Extract attributes and their values
attributes = set()
for entity in data:
    attributes.update(entity.keys())

# Prepare CSV rows
rows = []
for entity in data:
    row = [entity.get(attr, "") for attr in attributes]
    rows.append(row)

# Write data to CSV file
with open(filename, "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(list(attributes))  # Write header
    writer.writerows(rows)

print("Dataset downloaded and saved as CSV successfully!")
print('clinical_entity_dataset_{current_date}.csv')
```
